```python
SCHEMA = "workspace.sephora_products_and_skincare_reviews"
tables = (
    spark
    .sql(f"SHOW TABLES FROM {SCHEMA};")
    .select("tableName")
)
tables.show(n=1000, truncate=False)
```

```python
exceptions = {}
for row in tables.collect():
    table_name: str = row["tableName"]
    if not table_name.startswith("1_"):
        continue
    print(table_name)
    try:
        spark.sql(f"DROP TABLE IF EXISTS {SCHEMA}.{table_name};")
    except Exception as e:
        exceptions[table_name] = e

print("\nExceptions:")
print(exceptions)
```
