```python
SCHEMA = dbutils.widgets.get("SCHEMA")
VOLUME_DATA = dbutils.widgets.get("VOLUME_DATA")

TABLE_BRONZE_PRODUCTS_INFO = SCHEMA + "." + dbutils.widgets.get("TABLE_BRONZE_PRODUCTS_INFO")
TABLE_BRONZE_REVIEWS = SCHEMA + "." + dbutils.widgets.get("TABLE_BRONZE_REVIEWS")
```

```python
df_bronze_products_info = (
    spark
    .read
    .csv(
        f"{VOLUME_DATA}/product_info.csv",
        inferSchema=True,
        header=True
    )
    .select(
        "*",
        "_metadata.file_modification_time",
        "_metadata.file_path"
    )
)

(
    df_bronze_products_info
    .write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .format("delta")
    .saveAsTable(TABLE_BRONZE_PRODUCTS_INFO)
)
```

```python
df_bronze_reviews = (
    spark
    .read
    .csv(
        f"{VOLUME_DATA}/reviews*.csv",
        inferSchema=True,
        header=True
    )
    .select(
        "*",
        "_metadata.file_modification_time",
        "_metadata.file_path"
    )
)
(
    df_bronze_reviews
    .write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .format("delta")
    .saveAsTable(TABLE_BRONZE_REVIEWS)
)
display(df_bronze_reviews.limit(10))
```
