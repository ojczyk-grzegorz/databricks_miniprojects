```python
SCHEMA = "workspace.sephora_products_and_skincare_reviews"
tables = (
    spark
    .sql(f"SHOW TABLES FROM {SCHEMA};")
    .select("tableName")
    .collect()
)

for row in tables:
    table_name = row["tableName"]
    print(table_name)
    try:
        spark.sql(f"DROP TABLE IF EXISTS {table_name};")
    except Exception as e:
        print(e)
```
